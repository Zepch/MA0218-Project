```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
sb.set()
```

```python
df = pd.read_csv("jesterfinal151cols.csv")
df.fillna(99, inplace=True) # Change all null value into 99
df
```

# Data Cleaning

```python
count = [] # Create a list to see how much non null object in every columns
for i in df:
    if i == "Total_Joke": # We don't use the first column
        continue
    c = (df[i] != 99).sum()
    count.append(c)
 
# We then print all the important attributes
print("Statistics Summary:")
print(f"Mean               : {np.mean(count)}")
print(f"Median             : {np.median(count)}")
print(f"Standard deviation : {np.std(count)}")
print(f"Max                : {np.max(count)}")
print(f"Min                : {np.min(count)}")
```

### 1 Joke gets around 12,000 ratings
### Therefore, we candrop jokes that have less than 5000 ratings and the rest of null value will be filled by inputations

```python
df_new = df.copy() # Create a copy from the original data frame
for i in df:
    if i == "Total_Joke": # We drop total joke first
        df_new = df_new.drop(i, axis=1)
    c = (df[i] != 99).sum()
    if c < 5000:
        for j in range(0,50692): # Check each row
            value = df_new[i][j]
            if value != 99: # If the value of jth row on the coloumn that will be deleted is not a null value
                df['Total_Joke'][j] -= 1 # That specific row(user) total number of jokes will be reduced
        df_new = df_new.drop(i, axis=1)
df_new
```

### To fill the null values, the idea is we will use mean imputation by filling the mean rating of that user to the null value

```python
# First, we need to iterate each row, then replace the null value with the mean
for i, row in df_new.iterrows():
    mean = np.mean(row[row!=99])
    df_new.loc[i, 'Mean'] = mean
    df_new.iloc[i] = row.replace(99, mean)
df_new
```

```python
# We add back the total joke coloumn
df_new["Total_Joke"] = df["Total_Joke"]
df_new
```

```python
# Then, we can import back the new dataframe to a new csv file that can be used later in the Data Exploration
df_new.to_csv('new_data.csv', index=False)
```
